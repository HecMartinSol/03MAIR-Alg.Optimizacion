```
import time
# DECORADOR PARA MEDIR EL TIEMPO DE LOS ALGORITMOS
def timer_decorator(method):
    def measure_time(*args, **kw):
        time_begin = time.time()
        result = method(*args, **kw)
        time_end = time.time()
        ms_elapsed = (time_end - time_begin) * 1000

        if 'log_time' in kw:
            name = kw.get('log_name', method.__name__.upper())
            kw['log_time'][name] = ms_elapsed
        else:
            print(method.__name__, "lasted", ms_elapsed, "ms")

        return result
    return measure_time
```

```
# VUELTA ATRAS: problema de las reinas
def reinas(N, solucion, etapa):
    for i in range(1, N + 1):
        solucion[etapa] = i
        # print(solucion)

        if es_prometedor(solucion, etapa) : 
            if etapa == N - 1:
                print("Solucion posible para N =",N,":", solucion)
                return solucion
            else:
                # print("-> [P] prometedora")
                reinas(N, solucion, etapa + 1)
        else:
            # print("-> [N] no prometedora")
            pass
    
        solucion[etapa] = 0



def es_prometedor(solucion, etapa) : 
    # si la solucion tiene dos valores iguales no es valida
    for i in range(etapa + 1):
        if solucion.count(solucion[i]) > 1:
            return False

        # diagonales
        for j in range(i+1, etapa + 1):
            # print("comprueba diagonal de ", i , "y", j)
            if abs(i-j) == abs(solucion[i] - solucion[j]):
                return False
    
    return True

@timer_decorator 
def measure_queens(N, **kwargs):
  solution = []
  for i in range(N):
    solution.append(0)
  reinas(N, solution, 0)

for N in range(10):
  time_info ={"N":N}
  measure_queens(N, log_time=time_info)
  print(time_info)
```

#### PROBLEMA: ENCONTRAR LOS DOS PUNTOS MÁS CERCANOS
Dado un conjunto de puntos se trata de encontrar los dos puntos más cercanos

Guía para aprendizaje:
  - Suponer en 1D, o sea, una lista de números: [3403, 4537, 9089, 9746, 7259, ….
  
  
  
  
  
  
  
  
  
  
  
  
  
  
  
  
  
  - Primer intento: Fuerza bruta
  - Calcular la complejidad. ¿Se puede mejorar?
  - Segundo intento. Aplicar Divide y Vencerás
  - Calcular la complejidad. ¿Se puede mejorar?
  - Extender el algoritmo a 2D: [(1122, 6175), (135, 4076), (7296, 2741)…
  - Extender el algoritmo a 3D


```
import random
import math

MAX = 20
ELEMENTS = 10
LISTA_1D = [random.randrange(1,MAX) for x in range(ELEMENTS)]
LISTA_2D = [(random.randrange(1,MAX),random.randrange(1,MAX)) for x in range(ELEMENTS)]
LISTA_3D = [(random.randrange(1,MAX),random.randrange(1,MAX),random.randrange(1,MAX)) for x in range(ELEMENTS)]

# distancia entre dos puntos para 1D, 2D y 3D
def distance_points(a , b):
    # cálculo para tublas 2D o más
    if type(a) == tuple and type(b) == tuple:
        if len(a) != len(a):
            return -1

        sum_diff_squared_distances = 0
        for i in range(len(a)):
            sum_diff_squared_distances += ((a[i] - b[i]) ** 2)

        distance = math.sqrt(sum_diff_squared_distances)
    
    # cálculo para valores numéricos
    else:
        distance = abs(a - b)
        
    return distance

# Ordena la lista de elementos por x o por y
def ordenar_lista(theList, sort="x"):
  if type(theList[0]) == tuple:
    idx_sort = (1 if sort == "y" else 0)
    theList = sorted(theList, key=lambda x: x[idx_sort])
  else:
    theList = sorted(theList)
  
  return theList

#Elimina elementos duplicados en una lista
def eliminar_duplicados(lst):
    return list(set([i for i in lst])) 


# Calcula puntos mas cercanos por fuerza bruta
def puntos_mas_cercanos_fuerza_bruta(elements):
    minDist = math.inf
    elements = eliminar_duplicados(elements)
    for i in range(len(elements)):
        for j in range(i+1, len(elements)):
        
            p = elements[i]
            q = elements[j]

            currDist = distance_points(p, q)
            if currDist < minDist:
                minDist = currDist
                closestPair = {"pair":(p, q), "dist" : minDist}

    return closestPair


print("-----     FUERZA BRUTA     -----------------------")
print(LISTA_1D)
print(puntos_mas_cercanos_fuerza_bruta(LISTA_1D))
print("--------------------------------------------------")
# ---------------------------------------------------------
print(LISTA_2D)
print(puntos_mas_cercanos_fuerza_bruta(LISTA_2D))
print("--------------------------------------------------")
# ---------------------------------------------------------
print(LISTA_3D)
print(puntos_mas_cercanos_fuerza_bruta(LISTA_3D))
print("--------------------------------------------------")
# ---------------------------------------------------------
```

El problema puede ser resuelto en tiempo O(n log n) usando un algoritmo recursivo  tipo divide y vencerás, p. ej., como sigue:

  1. Ordenar los puntos según su coordenada X.
  2. Si el tamaño del conjunto es 2, devolver la distancia entre ellos. Si el conjunto tiene 0 o 1 elementos, devolver infinito.
  3. Dividir el conjunto de puntos en dos partes iguales (del mismo número de puntos).
  4. Solucionar el problema de forma recursiva en las partes izquierdas y derecha. 
      Esto devolverá una solución para cada parte, llamadas dLmin y dRmin. 
      Escoger el mínimo entre estas dos soluciones, llamado dLRmin.
  5. Seleccionar los puntos de la parte derecha e izquierda que están a una distancia horizontal menor que dLRmin de la recta divisoria entre ambos. 
      Aprovechar que los puntos están ordenados para elegir los últimos puntos de la parte izquierda y los primeros de la parte derecha.

  6. Encontrar la distancia mínima dCmin entre todos los pares de puntos formados por un punto de cada parte del paso anterior.
  7. La respuesta final es el mínimo entre dCmin y dLRmin.


```
import math
import random

#   Calculates the nearest points in 2D by brute force
def closest_pair_bruteforce(points_by_x):
    p1 = points_by_x[0]
    p2 = points_by_x[1]
    min_distance = euclidean_distance(p1, p2)

    length = len(points_by_x)
    if length == 2:
        return p1, p2, min_distance

    for i in range(length-1):
        for j in range(i + 1, length):
            if i != 0 and j != 1:
                d = euclidean_distance(points_by_x[i], points_by_x[j])
                if d < min_distance:
                    min_distance = d
                    p1, p2 = points_by_x[i], points_by_x[j]

    return p1, p2, min_distance

#   Calculates the distance between two points in 2D
def euclidean_distance(p1, p2):
    return math.sqrt((p1[0] - p2[0]) ** 2 + (p1[1] - p2[1]) ** 2)


def closest_split_pair(points_by_x, points_by_y, min_distance, best_pair):
    length_x = len(points_by_x)  # store length - quicker
    midpoint_x = points_by_x[length_x // 2][0]  # select midpoint on x-sorted array

    # Create points subarray of points not further than min_distance from midpoint on x-sorted array
    nearests_y = [x for x in points_by_y if midpoint_x - min_distance <= x[0] <= midpoint_x + min_distance]
    best_distance = min_distance  # assign best_distance value to min_distance
    length_nearests = len(nearests_y)  # store length of subarray for quickness
    
    for i in range(length_nearests - 1):
        for j in range(i+1, min(i + 7, length_nearests)):
            p, q = nearests_y[i], nearests_y[j]
            
            curr_distance = euclidean_distance(p, q)
            if curr_distance < best_distance:
                best_pair = p, q
                best_distance = curr_distance

    return best_pair[0], best_pair[1], best_distance

def closest_pair_recursive(points_by_x, points_by_y):
    length = len(points_by_x)
    if length <= 3:
        return closest_pair_bruteforce(points_by_x)

    middle = length // 2  # Division without remainder, need int
    left_points_x = points_by_x[:middle]  # Two-part split
    right_points_x = points_by_x[middle:]
    
    # Determine midpoint on x-axis
    qx = set(left_points_x)
    left_points_y = list()
    right_points_y = list()
    for x in points_by_y:
        if x in qx:
            left_points_y.append(x)
        else:
            right_points_y.append(x)
    
    # Call recursively both arrays after split
    (p1, q1, min_distance_left) = closest_pair_recursive(left_points_x, left_points_y)
    (p2, q2, min_distance_right) = closest_pair_recursive(right_points_x, right_points_y)

    # Determine smaller distance between points of 2 arrays
    if min_distance_left <= min_distance_right:
        min_distance = min_distance_left
        closests_points = (p1, q1)
    else:
        min_distance = min_distance_right
        closests_points = (p2, q2)
    
    # Call function to account for points on the boundary
    (p3, q3, mi3) = closest_split_pair(points_by_x, points_by_y, min_distance, closests_points)

    # Determine smallest distance for the array
    if min_distance <= mi3:
        return closests_points[0], closests_points[1], min_distance
    else:
        return p3, q3, mi3


def get_2d_tuple(length = 10000):
    lst1 = [random.randint(0, 1000) for i in range(length)]
    lst2 = [random.randint(0, 1000) for i in range(length)]
    points = list(zip(lst1, lst2))  # This produces list of tuples
    return points


points = get_2d_tuple(20)
points_by_x = sorted(points, key=lambda x: x[0])  # Presorting x-wise
points_by_y = sorted(points, key=lambda x: x[1])  # Presorting y-wise
p1, p2, min_distance = closest_pair_recursive(points_by_x, points_by_y)  # Recursive D&C function

print(points)
print(p1, p2)
print(min_distance)
```
