```
import time
# DECORADOR PARA MEDIR EL TIEMPO DE LOS ALGORITMOS
def timer_decorator(method):
    def measure_time(*args, **kw):
        time_begin = time.time()
        result = method(*args, **kw)
        time_end = time.time()
        ms_elapsed = (time_end - time_begin) * 1000

        if 'log_time' in kw:
            name = kw.get('log_name', method.__name__.upper())
            kw['log_time'][name] = ms_elapsed
        else:
            print(method.__name__, "lasted", ms_elapsed, "ms")

        return result
    return measure_time
```

```
# DIVIDE Y VENCERAS - LAS TORRES DE HANOI
def hanoi(fichas, origen, destino):
  if fichas == 1:
    print("Llevar la ficha de", origen, "a", destino)
  else:
    hanoi(fichas - 1, origen, 6 - origen - destino)
    print("Llevar la ficha de", origen, "a", destino)
    hanoi(fichas - 1, 6 - origen - destino, destino)


@timer_decorator 
def measure_hanoi(**kwargs):
  hanoi(5,1,3)

time_info ={}
measure_hanoi(log_time=time_info)

print(time_info)
```

```
# VUELTA ATRAS: problema de las reinas
def reinas(N, solucion, etapa):
    for i in range(1, N + 1):
        solucion[etapa] = i
        # print(solucion)

        if es_prometedor(solucion, etapa) : 
            if etapa == N - 1:
                print("Solucion posible para N =",N,":", solucion)
                return solucion
            else:
                # print("-> [P] prometedora")
                reinas(N, solucion, etapa + 1)
        else:
            # print("-> [N] no prometedora")
            pass
    
        solucion[etapa] = 0



def es_prometedor(solucion, etapa) : 
    # si la solucion tiene dos valores iguales no es valida
    for i in range(etapa + 1):
        if solucion.count(solucion[i]) > 1:
            return False

        # diagonales
        for j in range(i+1, etapa + 1):
            # print("comprueba diagonal de ", i , "y", j)
            if abs(i-j) == abs(solucion[i] - solucion[j]):
                return False
    
    return True

@timer_decorator 
def measure_queens(N, **kwargs):
  solution = []
  for i in range(N):
    solution.append(0)
  reinas(N, solution, 0)

for N in range(10):
  time_info ={"N":N}
  measure_queens(N, log_time=time_info)
  print(time_info)
```

```
"""
    PROBLEMA: ENCONTRAR LOS DOS PUNTOS MÁS CERCANOS
        Dado un conjunto de puntos se trata de encontrar los dos puntos más cercanos

    • Guía para aprendizaje:
        - Suponer en 1D, o sea, una lista de números: [3403, 4537, 9089, 9746, 7259, ….
        - Primer intento: Fuerza bruta
        - Calcular la complejidad. ¿Se puede mejorar?
        - Segundo intento. Aplicar Divide y Vencerás
        - Calcular la complejidad. ¿Se puede mejorar?
        - Extender el algoritmo a 2D: [(1122, 6175), (135, 4076), (7296, 2741)…
        - Extender el algoritmo a 3D
"""

import random
import math

MAX = 20
ELEMENTS = 10
LISTA_1D = [random.randrange(1,MAX) for x in range(ELEMENTS)]
LISTA_2D = [(random.randrange(1,MAX),random.randrange(1,MAX)) for x in range(ELEMENTS)]
LISTA_3D = [(random.randrange(1,MAX),random.randrange(1,MAX),random.randrange(1,MAX)) for x in range(ELEMENTS)]

def distance_points(a , b):
    # cálculo para tublas 2D o más
    if type(a) == tuple and type(b) == tuple:
        if len(a) != len(a):
            return -1

        sum_diff_squared_distances = 0
        for i in range(len(a)):
            sum_diff_squared_distances += ((a[i] - b[i]) ** 2)

        distance = math.sqrt(sum_diff_squared_distances)
    
    # cálculo para valores numéricos
    else:
        distance = abs(a - b)
        
    return distance

def sort_list(theList, sort="x"):
  if type(theList[0]) == tuple:
    idx_sort = (1 if sort == "y" else 0)
    theList = sorted(theList, key=lambda x: x[idx_sort])
  else:
    theList = sorted(theList)
  
  return theList
  

def closest_points_bruteforce(elements):
    minDist = math.inf
    for i in range(len(elements)):
        for j in range(i+1, len(elements)):
        
            p = elements[i]
            q = elements[j]

            currDist = distance_points(p, q)
            if currDist < minDist:
                minDist = currDist
                closestPair = {"pair":(p, q), "dist" : minDist}

    return closestPair

print("-----     FUERZA BRUTA     -----------------------")
print(LISTA_1D)
print(closest_points_bruteforce(LISTA_1D))
print("--------------------------------------------------")
# ---------------------------------------------------------
print(LISTA_2D)
print(closest_points_bruteforce(LISTA_2D))
print("--------------------------------------------------")
# ---------------------------------------------------------
print(LISTA_3D)
print(closest_points_bruteforce(LISTA_3D))
print("--------------------------------------------------")
# ---------------------------------------------------------


"""
El problema puede ser resuelto en tiempo O(n log n) usando un algoritmo recursivo  tipo divide y vencerás, p. ej., como sigue:

  1. Ordenar los puntos según su coordenada X.
  2. Si el tamaño del conjunto es 2, devolver la distancia entre ellos. Si el conjunto tiene 0 o 1 elementos, devolver infinito.
  3. Dividir el conjunto de puntos en dos partes iguales (del mismo número de puntos).
  4. Solucionar el problema de forma recursiva en las partes izquierdas y derecha. 
      Esto devolverá una solución para cada parte, llamadas dLmin y dRmin. 
      Escoger el mínimo entre estas dos soluciones, llamado dLRmin.
  5. Seleccionar los puntos de la parte derecha e izquierda que están a una distancia horizontal menor que dLRmin de la recta divisoria entre ambos. 
      Aprovechar que los puntos están ordenados para elegir los últimos puntos de la parte izquierda y los primeros de la parte derecha.

  6. Encontrar la distancia mínima dCmin entre todos los pares de puntos formados por un punto de cada parte del paso anterior.
  7. La respuesta final es el mínimo entre dCmin y dLRmin.
"""

def closest_split_pair(p_x, p_y, delta, best_pair):
    ln_x = len(p_x)  # store length - quicker
    mx_x = p_x[ln_x // 2][0]  # select midpoint on x-sorted array
    # Create a subarray of points not further than delta from
    # midpoint on x-sorted array
    s_y = [x for x in p_y if mx_x - delta <= x[0] <= mx_x + delta]
    best = delta  # assign best value to delta
    ln_y = len(s_y)  # store length of subarray for quickness
    for i in range(ln_y - 1):
        for j in range(i+1, min(i + 7, ln_y)):
            p, q = s_y[i], s_y[j]
            dst = dist(p, q)
            if dst < best:
                best_pair = p, q
                best = dst
    return best_pair[0], best_pair[1], best


def closest_points_recursive(elementsByX, elementsByY):
    length = len(elementsByX)
    # si el tamaño es menor que 3 se puede resolver por fuerza bruta
    if length <= 3:
        return closest_points_bruteforce(elementsByX)

    # Se divide en dos mitades
    middle = length // 2
    left_X = elementsByX[:middle]
    right_X = elementsByX[middle:]

    print(middle, left_X, right_X)

    # Obtenemos el elemento en el punto medio
    if type(elementsByX[0]) == tuple :
      middlepoint = elementsByX[middle][0]
    else:
      middlepoint = elementsByX[middle]

    print(middlepoint)

    # Dividimos el array ordenado por Y por distancias mas cercanaas
    left_Y = list()
    right_Y = list()
    for x in elementsByY:
        if x[0] <= middlepoint:
           left_Y.append(x)
        else:
           right_Y.append(x)

    print(left_X, left_Y)
    print(right_X, right_Y)

    # Resolvemos recursivamente ambas partes
    (p1, q1, min_distance_left) = closest_points_recursive(left_X, left_Y)
    (p2, q2, min_distance_right) = closest_points_recursive(right_X, right_Y)

    print(p1, q1, min_distance_left)
    print(p2, q2, min_distance_right)

    # determinar cuales tienen menor distancia
    if min_distance_left <= min_distance_right:
        min_dist = min_distance_left
        min_dist_points = (p1, q1)
    else:
        min_dist = min_distance_right
        min_dist_points = (p2, q2)

      
    # Call function to account for points on the boundaright_Y
    (p3, q3, min_distance_final) = closest_split_pair(elementsByX, elementsByY, min_dist, min_dist_points)


    # min_distetermine smallest min_dististance for the arrelementsByY
    if min_dist <= min_distance_final:
        return min_dist_points[0], min_dist_points[1], min_dist
    else:
        return p3, q3, min_distance_final
    

list_by_x = sort_list(LISTA_2D, "x")
list_by_y = sort_list(LISTA_2D, "y")
solution = closest_points_recursive(list_by_x, list_by_y)
print(solution)
    
```
