````
# INICIALIZACION DE LAS ESTRUCTURAS
ciphers = [1,2,3,4,5,6,7,8,9]
operations = ["+","-","*","/"]

# DECORADOR PARA MEDIR EL TIEMPO DE LOS ALGORITMOS
import time
def timer_decorator(method):
    def measure_time(*args, **kw):
        time_begin = time.time()
        result = method(*args, **kw)
        time_end = time.time()
        ms_elapsed = (time_end - time_begin) * 1000

        if 'log_time' in kw:
            name = kw.get('log_name', method.__name__.upper())
            kw['log_time'][name] = ms_elapsed
        else:
            print(method.__name__, "lasted", ms_elapsed / 1000, "s")

        return result
    return measure_time
````

## Diseña un algoritmo para resolver el problema por fuerza bruta

````
import random as random

@timer_decorator
def bruteforce(ciphers, operations, n):
  solutions = []

  for x1 in ciphers:
    ciphers_1 = ciphers[:]
    ciphers_1.remove(x1)

    for op1 in operations:
      operations_1 = operations[:]
      operations_1.remove(op1)

      for x2 in ciphers_1:
        ciphers_2 = ciphers_1[:]
        ciphers_2.remove(x2)

        for op2 in operations_1:
          operations_2 = operations_1[:]
          operations_2.remove(op2)

          for x3 in ciphers_2:
            ciphers_3 = ciphers_2[:]
            ciphers_3.remove(x3)

            for op3 in operations_2:
              operations_3 = operations_2[:]
              operations_3.remove(op3)

              for x4 in ciphers_3:
                ciphers_4 = ciphers_3[:]
                ciphers_4.remove(x4)

                for op4 in operations_3:
                  operations_4 = operations_3[:]
                  operations_4.remove(op4)

                  for x5 in ciphers_4:
                    ciphers_5 = ciphers_4[:]
                    ciphers_5.remove(x5)

                    expression = str(x1) + str(op1) + str(x2) + str(op2) + str(x3) + str(op3) + str(x4) + str(op4) + str(x5)
                    if eval(expression) == n:
                      solutions.append(expression)
  return solutions

n = 4 
solutions = bruteforce(ciphers, operations, n)
print(len(solutions), "soluciones posibles para calular el valor", n)
print(solutions)
````

## Calcula la complejidad del algoritmo por fuerza bruta

Siendo n el tamaño de la lista de números disponibles, y siendo m el tamaño de la lista de las operaciones disponibles; el algoritmo tiene 5 recorridos de n y 4 de m de la siguiente manera:

$n*m*(n-1)*(m-1)*(n-2)*(m-2)*(n-3)*(m-3)*(n-4)$

Generalizando para cualquier numero tenemos:<br>
$ \frac{n! * m!}{(n-m)!} \ |\ n > m\ \forall n,m \geq 1 $

Por lo que podemos simplificando podemos decir que el algoritmo es<br>
$O(n) \simeq n! $

Para encontrar la frontera de soluciones de problema buscaremos soluciones del problema para todo los rangos de valores.

1. Se explorarán todos los valores positivos del 0 al 100 y se almacenará en un diccionario
  - Tiempo estimado: $100\ numeros * 3.5 s = 350\ seg. \simeq 6 min$
2. Se explorarán todos los valores positivos del 0 al -100 y se almacenará en un diccionario
  - Tiempo estimado: $100\ numeros * 3.5 s = 350\ seg. \simeq 6 min$
3. Se recorrerá para comprobar cuales tienen soluciones y cual no

````
# Resolución del problema de 0 a 100
max_estimated = 100
positives = {}
for i in range(max_estimated):
  solutions = bruteforce(ciphers, operations, i)
  random_choice = random.choice(solutions) if len(solutions) > 0 else None
  positives[i] = (len(solutions), random_choice)
````

````
# Resolución del problema de -100 a 0
min_estimated = -100
negatives = {}
for i in range(min_estimated, 0):
  solutions = bruteforce(ciphers, operations, i)
  random_choice = random.choice(solutions) if len(solutions) > 0 else None
  negatives[i] = (len(solutions), random_choice)
````

````
# Comprobación de fronteras
for x in positives:
  if (positives[x][0] == 0):
    print(x, "no tiene soluciones")
for x in negatives:
  if (negatives[x][0] == 0):
    print(x, "no tiene soluciones")
````

## (*)Diseña un algoritmo que mejore la complejidad del algortimo por fuerza bruta. Argumenta porque crees que mejora el algoritmo por fuerza bruta

Dados los datos obtenidos en las pruebas de los límites anteriores, podemos hacer un algoritmo de marcha atrás que descarte algunas soluciones.

Una solucion no será prometedora si:
- No quedan operaciones por hacer
- El resultado actual es menor o igual que -69 y solo hay operaciones de - o /
- El resultado actual es mayor o igual que +77 y solo hay operaciones de + o *

````
def isSolution(n, expression, operations):
  return len(operations) == 0 and eval(expression) == n

def isPromising(n, expression, operations):
  # print("isPromising", n, expression, operations)
  operationsRemaining = len(operations) >= 0
  currVal = eval(expression)
  upLimit = currVal < 77 or (currVal >= 77 and ("-" in operations or "/" in operations))
  downLimit = currVal > -69 or (currVal <= -69 and ("+" in operations or "*" in operations))
  # print("operationsRemaining", operationsRemaining, "upLimit", upLimit, "downLimit", downLimit)
  allOk = operationsRemaining and upLimit and downLimit
  return allOk

def backtrack_numbers(n, expression, ciphers, operations, step, solutions):
  for op in operations:
    step_operations = operations[:]
    step_operations.remove(op)

    for c in ciphers:
      step_ciphers = ciphers[:]
      step_ciphers.remove(c)
    
      step_expression = str(expression) + str(op) + str(c)

      if isPromising(n, step_expression, step_operations): 
        # print("[",step,"]", step_expression, "=", eval(step_expression), " => prometedor", str(step_ciphers), str(step_operations))
        if isSolution(n, step_expression, step_operations):
          # print("[",step,"]", step_expression, "=", eval(step_expression), " => SOLUCION!")
          solutions.append(step_expression)
        else:
          backtrack_numbers(n, step_expression, step_ciphers, step_operations, step+1, solutions)
      else:
        # print("[",step,"]", step_expression, "=", eval(step_expression), " => no es prometedor")
        pass

@timer_decorator
def backtracking(n, ciphers, operations):
  solutions = []
  for i in ciphers:
    ciphers_i = ciphers[:]
    ciphers_i.remove(i)

    backtrack_numbers(n, str(i), ciphers_i, operations, 0, solutions)
  return solutions

n = 4
solutions = backtracking(n, ciphers, operations);
print(len(solutions), "soluciones posibles para calular el valor", n)
print(solutions)
````

## (*)Calcula la complejidad del algoritmo 

Pese a que la implementación del algoritmo es diferente mediante recursividad, el espacio de soluciones se tiene que recorrer es prácticamente completo ya que es necesario probar todas las combinaciones.<br>
La única diferencia es que en lugar de tener todos los bucles se realiza mediante recursividad.<br>
No obstante, la complejidad se mantiene como por fuerza bruta:<br>
$O(n) \simeq n! $

También observamos que tarda incluso más, probablemente debido a las comprobaciones adicionales que se realiza en la función de factibilidad que hacen que ralentize el proceso al evaluar la expresión.
